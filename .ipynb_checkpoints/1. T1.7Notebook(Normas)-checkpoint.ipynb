```python
# FCD's (Funciones de Conjuntos Difusos)
# a: amplitud
# b: ancho de la campana
# c: desplazamiento
def campana(x, a = 1, b = 3, c = 0):
    return 1/(1 + abs((x-c)/a)**(2*b))

# c: desplazamiento
# sigma: esperanza matematica
def gaussiana(x, c = 0, sigma = 1):
    return math.e ** (- 1/2 * ((x-c)/sigma)**2 )
```

# T-Normas

```python
def tnorma_dombi(a, b, p_lambda = 1):
    return 1 / ( 
        1 + ( (1/a - 1)**p_lambda + (1/b - 1)**p_lambda ) ** (1/p_lambda) 
    )

def tnorma_frank(a, b, s = 0.5):
    return np.emath.logn(s, (
        1 + (((s ** (1-a) - 1) * (s ** (1-b) - 1)) / (s-1))
    ))

def tnorma_hamacher(a, b, r = 1):
    return a * b / (
        r + (1 - r)*(a + b - a*b)
    )
```

# S-Normas

```python
# lambda > 0
def snorma_dombi(a, b, p_lambda = 1):
    return 1 / ( 
        1 + ( (1/a - 1)**p_lambda + (1/b - 1)**p_lambda ) ** (-1/p_lambda) 
    )

# s > 0, s != 1
def snorma_frank(a, b, s = 0.5):
    return 1 - np.emath.logn(s, (
        1 + (((s ** (1-a) - 1) * (s ** (1-b) - 1)) / (s-1))
    ))

# r > 0
def snorma_hamacher(a, b, r = 1):
    return (a + b + (r - 2) * a*b) / (r + (r - 1) * a*b)
```

## Dombi

```python
plt.title('S-Norma Dombi')
plotAB()

c1 = snorma_dombi(a, b)
c2 = snorma_dombi(a, b, p_lambda=2)
c3 = snorma_dombi(a, b, p_lambda=4)
plotC1('lambda = 1')
plotC2('lambda = 2')
plotC3('lambda = 4')
```

## Frank

```python
plt.title('S-Norma Frank')
plotAB()

c1 = snorma_frank(a, b)
c2 = snorma_frank(a, b, s=0.01)
c3 = snorma_frank(a, b, s=8)
plotC1('s = 0.5')
plotC2('s = 0.01')
plotC3('s = 8')
```
