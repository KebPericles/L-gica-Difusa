```python
## Proponga un universo de discurso X y a 2 conjuntos difusos A y B con cierto traslape entre ellos.
#Usando el lenguaje de programaci√≥n Python obtenga:
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x0 = 0
xd = 100
xs = xd/200

X = np.arange(x0,xd + xs,xs)
```

```python
# FUNCIONES DE CONJUNTOS DIFUSOS

def triangulo(x, xa, xb, creciente):
    m = 1 / (xb - xa)
    cb_xa = 0
    cb_xb = 1
    cb_xcentrado = lambda x: m * (x - xa)
    if not creciente:
        m *= -1
        cb_xa = 1
        cb_xb = 0
        cb_xcentrado = lambda x: m * x - m * xb

    return np.piecewise(x, [x <= xa, x >= xb], [cb_xa, cb_xb, cb_xcentrado])

t = lambda x: triangulo(x, 0, 35, True)
```

```python
# Imprimir ab y otro

def printABC(a, b, c, str):
    plt.plot(X, c, color='red', linewidth=4, label=str)
    plt.plot(X, a, '--', color='mediumpurple', label='Conjunto A')
    plt.plot(X, b, '--', color='blue', label='Conjunto B')
    
    plt.xlabel('x')
    plt.ylabel('\u03bc')
    plt.title(str)
    plt.legend(loc='best')
    plt.grid(True)
```

```python
A = t(X-20)
B = 1-t(X-30)
printABC(A, B, np.ones(201), 'Universo')
```

```python
# Uni√≥n de A y B
union = np.maximum(A,B)
printABC(A,B,union, 'Uni√≥n de A y B')
```

```python
# Intersecci√≥n de A y B
inter = np.minimum(A,B)
printABC(A,B,inter, 'Intersecci√≥n de A y B')
```

```python
# Complemento de A
compa = 1 - A
printABC(A,B,compa, 'Complemento de A')
```

```python
# Diferencia de A y B
diferencia = np.minimum(A, 1-B)
printABC(A,B,diferencia, 'Diferencia de A y B')
```
