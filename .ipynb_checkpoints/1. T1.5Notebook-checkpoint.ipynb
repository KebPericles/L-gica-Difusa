```python
# Librerías
import numpy as np
from matplotlib import pyplot as plt

# Imprimir
def plotArray(array, color, label, lw = 1, ls = "-"):
    plt.plot(x, array, color=color, label=label, linewidth=lw, linestyle=ls)
    
    plt.xlabel('x')
    plt.ylabel('\u03bc')
    plt.legend(loc='best')
    plt.grid(True)
```

## Ejercicio 2

```python
# Funciones imprimir
def plotA():
    plotArray(a, 'mediumpurple', 'Transistor')

def plotB():
    plotArray(b, 'blue', 'Resistor')

def plotAB():
    plotA()
    plotB()

def plotC(str = ''):
    plotArray(c, 'orange', str, lw=1.5, ls='-.')

def plot(title=''):
    plotAB()
    plotC(title)
```

```python
# Conjuntos
x = np.array([0,1,2,3,4,5])

a = mu_trans = np.array([0,0.3,0.7,0.8,0.9,1])
b = mu_res = np.array(  [0,0.2,0.4,0.6,0.8,1])

plotAB()
```

### 1) Union

```python
c = union = np.maximum(mu_trans, mu_res)
plot('Union')
```

### 2) Intersección

```python
c = inter = np.minimum(mu_trans,mu_res)
plot('Intersección')
```

### 3) Complemento $\mu_T$

```python
c = comp_trans = 1 - mu_trans
plotA()
plotC('Complemento')
```

### 4) Complemento $\mu_R$

```python
c = comp_res = 1 - mu_res
plotB()
plotC('Complemento')
```

### 5) Complemento de la intersección

```python
c = comp_inter = 1 - inter
plot('Complemento de la intersección')
```

## Ejercicio 3

```python
# Funciones imprimir
def plotA():
    plotArray(a, 'b', 'Conjunto A')

def plotB():
    plotArray(b, 'g', 'Conjunto B')

def plotC():
    plotArray(c, 'm', 'Conjunto C')

def plotD(str = ''):
    plotArray(d, 'orange', str, lw=1.5, ls='-.')

def plotE(str = ''):
    plotArray(e, 'c', str, lw=1.5, ls='-.')

def plotF(str = ''):
    plotArray(f, 'r', str, lw=1.5, ls='-.')
```

```python
# Universo
x0 = 0
xd = 10
xs = (x0 + xd) / 200

x = np.arange(x0, xd + xs, xs)
```

```python
# Conjuntos
a = mu_a = x / (x + 2)
b = mu_b = 2 ** (-x)
c = mu_c = 1/(1+10*(x-2)**2)

plotA()
plotB()
plotC()
```

### 1) Complementos

```python
d = comp_a = 1 - mu_a
e = comp_b = 1 - mu_b
f = comp_c = 1 - mu_c
```

```python
plotA()
plotD()
```

```python
plotB()
plotE()
```
