```python
import numpy as np
import math
from matplotlib import pyplot as plt

x = np.arange(-5, 5.1, 0.1)
```

## Discretas

### Triangulo

```python
# TODO Parametrizar los intervalos

def triangulo(x):
    tmp_triangulo = np.piecewise(x, [x < 0, x >= 0], [lambda x: x + 1, lambda x: -x + 1])
    return np.piecewise(tmp_triangulo, [tmp_triangulo < 0, tmp_triangulo > 0], [0, lambda x: x])

test_triangulo = triangulo(x/3)
plt.plot(x, test_triangulo)
```

### Trapecio

```python
# TODO Parametrizar los intervalos

def trapecio(x):
    tmp_trapecio = np.piecewise(x,
                                [x < -1.5,          x > -1.5, x > -0.5,            x > 0.5, x > 1.5],
                                [       0, lambda x: x + 1.5,        1, lambda x: -x + 1.5,       0])
    return tmp_trapecio

test_trapecio = trapecio(x)
plt.plot(x, test_trapecio)
```

## Continuas

### Gaussiana

```python
# c: desplazamiento
# sigma: esperanza matematica
def gaussiana(x, c = 0, sigma = 1):
    return math.e ** (- 1/2 * ((x-c)/sigma)**2 )

test_gaussiana = gaussiana(x)
plt.plot(x, test_gaussiana)
```

### Campana

```python
# a: amplitud
# b: ancho de la campana
# c: desplazamiento
def campana(x, a = 1, b = 3, c = 0):
    return 1/(1 + abs((x-c)/a)**(2*b))

test_campana = campana(x)
plt.plot(x, test_campana)
```

### Sigmoide

```python
# a: pendiente
# c: desplazamiento
def sigmoide(x, a = 2, c = 0):
    return 1 / (1 + math.e ** (-a * (x - c)))

test_sigmoide = sigmoide(x)
plt.plot(x, test_sigmoide)
```
